## Literature Review (with dataset/selection details)

### Albritton et al. 2018 — *Health Affairs*
- **Question:** Did HRRP reduce 30-day HF readmissions and were reductions offset by observation stays?
- **Data & Filters:** Medicare FFS 2006–2014 (HF vs non-targets). Excluded small-volume hospitals and those with **<5 observation stays** to stabilize estimates.
- **Method:** DiD for readmit vs observation trends.
- **Findings:** Readmissions fell, observation rose nearly 1-for-1 → total acute use ≈ unchanged.
- **Implication:** HRRP improved metrics more than utilization; outcomes (mortality/OOP) not tied to substitution.

### Sheehy et al. 2021 — *Journal of Hospital Medicine*
- **Question:** Did HRRP increase observation hospitalizations for targeted conditions?
- **Data & Filters:** Medicare FFS 2010–2016. Targeted vs non-targeted; dropped **obs <8h**; excluded unreliable billers.
- **Method:** Retrospective cohort; Poisson w/ robust SE.
- **Findings:** Observation ↑ for targeted conditions; readmit+obs combined ≈ unchanged.
- **Implication:** Site-of-care shift, not demand reduction.

### Wright et al. 2023 — *Health Services Research*
- **Question:** Long-run substitution into observation?
- **Data & Filters:** Medicare FFS 2010–2018; **7,162,189** stays (~18% targeted). Dropped hospitals with **<5 obs/year** and **obs <6h**.
- **Method:** ITS with hospital & time FE.
- **Findings:** Targeted conditions: obs **+0.02 pp/month** post-HRRP (95% CI 0.002, 0.04); COPD NS.
- **Implication:** Small monthly shifts accumulate; substitution persists without net acute-use decline.

### Qiu et al. 2022 — *Production & Operations Management*
- **Question:** Did readmit reductions come without mortality harm?
- **Data & Filters:** Medicare FFS 2009–2016; HF/AMI/pneumonia; excluded hospitals with **<100 target admissions/year**; continuous A/B; age ≥65.
- **Method:** Mortality hazards + 2-stage regressions linking readmit changes to mortality.
- **Findings:** Readmits ↓; mortality not improved and in some settings ↑ (aligning with JAMA/BMJ estimates **+0.22–0.40 pp** 30-day mortality).
- **Implication:** Avoiding necessary readmits can be harmful.

### Rogstad et al. 2022 — Systematic Review (social risk)
- **Question:** Are penalties inequitable?
- **Data & Filters:** 50+ empirical studies (2006–2021) explicitly including SES factors (race, dual-eligibility, disability).
- **Method:** Narrative synthesis.
- **Findings:** 7 studies: SES adjustment reduces penalty gaps; 3: peer-grouping reduces safety-net penalties.
- **Implication:** Lack of social-risk adjustment amplifies inequities.

### Albritton et al. 2017 — *eGEMS*
- **Question:** Did HRRP encourage longer observation **duration** near day-30?
- **Data & Filters:** Medicare 2006–2013 (HF). Excluded **obs <24h**; hospitals with **<25 HF readmits/year** or **<5 obs**.
- **Method:** RD on days-since-discharge; donut RD; optimal bandwidth.
- **Findings:** Overall null; **non-penalized hospitals** show longer obs just past 30 days.
- **Implication:** Duration manipulation is selective; penalty context matters.

### Nuckols et al. 2017 — *JAMA Internal Medicine*
- **Question:** Which interventions are cost-effective to reduce readmits?
- **Data & Filters:** RCTs/quasi-experiments **with cost data** and 30-day endpoints.
- **Method:** Systematic review/meta-analysis.
- **Findings:** Transitional care & patient-engagement often effective; savings mixed.
- **Implication:** Penalties alone underperform vs proactive care models.

### Kirsch et al. 2024 — Safety-net Clinician Survey
- **Question:** Workforce effects of HRRP?
- **Data & Filters:** ~300 clinician surveys (2020–2022); hospitals with **<5 respondents** excluded.
- **Method:** Mixed methods (survey + qualitative coding).
- **Findings:** Stress, morale loss, resource strain in safety-net settings.
- **Implication:** Penalties spill over onto workforce sustainability.

## Research Question
**To what extent did hospitals’ substitution of inpatient readmissions into observation stays under HRRP increase Medicare beneficiaries’ financial and welfare burdens—measured by out-of-pocket (OOP) spending, skilled nursing facility (SNF) access, and mortality—and how do these effects differ for penalized vs. unpenalized hospitals?**

---

## Abstract
HRRP modestly reduced readmissions but spurred substitution into **observation stays** (e.g., Wright 2023: +0.02 pp/month; Albritton 2018: +5.4% pre/post-30-day cliff among non-penalized hospitals). Observation is billed under Part B, exposing beneficiaries to higher OOP costs and **SNF ineligibility** (three-day inpatient rule). Penalties fall more heavily on safety-net hospitals, and several studies suggest mortality may worsen for certain conditions.  
This study centers observation in HRRP evaluation via a hospital-level **Observation Substitution Index (OSI)**, then quantifies how **penalized vs. unpenalized** status moderates the relationship between substitution and patient harm (OOP, SNF access, mortality). Using Medicare claims (2008–2020), I deploy event studies and difference-in-differences with hospital and time FE, plus robustness with RD around day-30 and IV using **expected penalty pressure**. Findings will inform whether CMS should include observation in quality metrics and revise SNF eligibility rules.

## Introduction (condensed)
- HRRP (FY2013) penalizes excess 30-day readmissions (initially HF, AMI, pneumonia; later COPD, joint replacement).
- Evidence indicates substitution **from readmits → observation** without reducing total acute revisits; mortality and equity concerns persist.
- **Gap:** Patient-level **financial and access harms** from substitution remain under-measured; penalty context (penalized vs unpenalized) may shape strategies and impacts.

---

## Conceptual Flow
HRRP penalties  
→ Hospitals reduce measured readmits  
→ **Substitute** to observation (and sometimes ED)  
→ Patients face **higher OOP**, **lost SNF coverage**, potential **outcome harms**  
→ Effects may differ by **penalty status** (penalized vs unpenalized) and **safety-net** status.

---

## Hypotheses
1. **Substitution & Costs:** Higher-OSI hospitals exhibit larger **post-HRRP increases in OOP** per patient.  
2. **SNF Access:** Higher OSI is associated with **lower probability of covered SNF** use within 30 days.  
3. **Mortality:** Higher OSI is associated with **higher 30-/90-day mortality**.  
4. **Penalty Heterogeneity:** Relationships in H1–H3 are **stronger among penalized hospitals** and/or hospitals with higher **penalty intensity**.

## Measures

### Observation Substitution Index (OSI)
$$
OSI_h = \frac{\Delta \text{Observation}_h}{|\Delta \text{Readmission}_h|}
$$

- Changes are measured **per 100 index discharges** (pre vs post HRRP).  
- **Variants:**  
  - Duration-weighted OSI: weight obs by **hours** (HCPCS G0378 units).  
  - Cost-weighted OSI: weight obs by **beneficiary liability** within 30 days.  
  - Outcome-weighted OSI: weight by **mortality risk** of the return modality.

### Penalty Status/Intensity
- **Penalized\_{hy}:** indicator if hospital $h$ receives an HRRP penalty in year $y$ (>0%).  
- **PenaltyRate\_{hy}:** continuous penalty percentage (0–3%).  
- **ExpectedPenalty\_{hy} (instrument):** baseline (pre-policy) target-condition share × national HRRP schedule.

## Data & Construction

### Sources (2008–2020)
- **MedPAR (IP):** index discharges, 30-day readmissions (exclude planned).  
- **Outpatient (OP):** observation (rev **0762**, HCPCS **G0378/G0379**), duration = G0378 units; ED (rev 0450–0459).  
- **Carrier/Part B:** patient liabilities (OOP).  
- **SNF claims:** admissions, coverage (3-day rule).  
- **MBSF:** demographics, dual-eligibility, death dates.  
- **HRRP Penalties:** annual penalty %, targeted conditions.  
- **AHA/HCRIS:** ownership, size, teaching status, financials.  
- **ADI/SVI:** area-level SES controls.

### Key Cohort Rules
- Age ≥65; continuous A & B coverage.  
- Exclude in-hospital deaths on index stay.  
- Exclude transfers.  
- Drop hospitals with **<5 obs/year**; drop obs stays <6–8h.  
- Initial focus: HF, AMI, pneumonia.  
- Define **Any Acute Return (30d)** = readmit ∪ obs ∪ ED.

## Identification & Estimation

### Baseline DiD with Penalty Layer
$$
Y_{iht} = \alpha + \beta \,(Post_t \times HighOSI_h) + 
\theta \,(Post_t \times HighOSI_h \times Penalized_{hy}) +
\gamma X_{iht} + \mu_h + \lambda_t + \epsilon_{iht}
$$

- $Y_{iht}$: OOP (30d), SNF access, Mortality (30/90d).  
- $X_{iht}$: patient controls.  
- $\mu_h$: hospital FE.  
- $\lambda_t$: year FE.  
- $\beta$: average HighOSI effect.  
- $\theta$: incremental effect among penalized hospitals.

### Event Study
$$
Y_{iht} = \alpha + \sum_{k \neq -1}\beta_k \,\mathbb{1}[EventTime=k] \times HighOSI_h
+ \sum_{k \neq -1}\theta_k \,\mathbb{1}[EventTime=k] \times HighOSI_h \times Penalized_{hy}
+ \gamma X_{iht} + \mu_h + \lambda_t + \epsilon_{iht}
$$

### IV Robustness
- **First stage:** substitution intensity on ExpectedPenalty.  
- **Second stage:** outcomes on predicted substitution.

### RD Robustness
- Running variable: days since discharge.  
- Cutoff: 30 days.  
- Outcomes: probability(obs), obs hours.  
- Stratify by penalty status.

## Threats to Validity
- Coding drift → comorbidity controls, FE.  
- Concurrent policies (Two-Midnight Rule) → year/policy indicators.  
- Case mix → patient-level controls.  
- Small hospitals → exclusions.  
- Non-parallel pre-trends → event-study checks.  
- Spillovers → HRR FE in sensitivity.

---

## Expected Contribution
- **Measurement innovation:** OSI as quantifiable substitution metric.  
- **Welfare quantification:** Converts “metric gaming” into dollars (OOP), access (SNF), and health (mortality).  
- **Penalty heterogeneity:** Shows penalized hospitals substitute more, harming patients more.  
- **Policy relevance:** Evidence for counting obs in HRRP metrics, revising SNF rule, and adjusting for social risk.
