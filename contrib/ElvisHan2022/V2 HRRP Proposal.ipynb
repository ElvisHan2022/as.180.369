```python
# Behavioral Adaptation under the Hospital Readmissions Reduction Program (HRRP): Substitution, Upcoding, and Timing Responses

---

## Abstract

The Hospital Readmissions Reduction Program (HRRP), introduced under the Affordable Care Act in 2013, penalizes hospitals with excess 30-day readmissions for select conditions to promote value-based care. Early studies posited that hospitals **“gamed” the system** by shifting returning patients from inpatient readmissions (penalized) to outpatient **observation stays** (unpenalized). However, recent analyses—including my own examination of Medicare Part A and B data—show that **observation stays actually decreased** in the aggregate post-HRRP.  

This finding raises a central question: *If hospitals were not substituting into observation stays, how did they adapt to HRRP incentives?*  

This project re-examines HRRP’s behavioral effects by comparing three potential mechanisms of hospital response:

1. **Substitution** — reclassification of readmissions into observation stays.  
2. **Upcoding** — inflating case-mix severity (higher DRG weights) to reduce apparent readmission risk.  
3. **Timing Shifts** — delaying readmissions beyond 30 days or increasing early post-discharge mortality to avoid penalties.

Using linked **MedPAR**, **Outpatient**, and **MBSF** data (2008–2020) with hospital-level penalty exposure, I will quantify how each mechanism evolved and whether these adaptations differed across **safety-net and non-safety-net hospitals**.  

By disentangling behavioral channels, this study provides a more accurate understanding of how hospitals internalize incentive-based regulation—clarifying whether HRRP improved efficiency or merely altered reporting behavior.

---

## 1. Introduction

### 1.1 Policy Context

The Hospital Readmissions Reduction Program (HRRP) ties Medicare reimbursement to hospitals’ 30-day readmission performance for specific conditions such as heart failure, acute myocardial infarction, and pneumonia. Hospitals exceeding national averages face up to a 3% reduction in total Medicare payments.  

This design created strong incentives to reduce measured readmissions—but **only within the 30-day inpatient window**, leaving open alternative ways to avoid penalties.  

### 1.2 The Puzzle

Prevailing literature argues that hospitals responded by increasing **observation stays**—short outpatient admissions not counted as readmissions (Gupta 2018; Sheehy 2021).  
Yet, new descriptive analysis using Medicare Part A and B data from 2008–2020 reveals that **observation utilization actually fell** over time, contradicting prior claims.  

This empirical puzzle suggests that hospitals may have adapted through other, less visible channels such as **coding adjustments, discharge timing manipulation, or triage changes**, rather than simple substitution.  

### 1.3 Research Questions

This study investigates the full set of behavioral margins available to hospitals under HRRP:

1. Did hospitals **substitute** readmissions with observation stays, or did overall post-discharge encounters decline?  
2. Did hospitals **upcode** DRG severity to appear compliant while offsetting penalties?  
3. Did hospitals **shift timing**, delaying readmissions or increasing early mortality?  
4. How did these responses vary between **safety-net** and **non-safety-net** hospitals?

### 1.4 Conceptual Framework

Hospitals maximize net revenue under HRRP subject to constraints on quality and resource availability. When penalized, they may respond by:
- **Reclassification** (substitution);
- **Coding adjustment** (administrative inflation);
- **Temporal shifting** (postponing readmissions beyond penalty window).  

Each channel represents an attempt to reduce measured readmissions without necessarily improving patient outcomes.  

---

## 2. Literature Review (Condensed)

- **Substitution:** Early HRRP studies (Zuckerman 2016; Gupta 2018) found increased observation stays, though later evidence (Yakusheva 2020; Wright 2023) shows these effects are smaller and inconsistent.  
- **Upcoding:** Ibrahim et al. (2018) estimate that up to two-thirds of reported readmission reductions may reflect administrative risk inflation rather than genuine improvement.  
- **Timing:** Gupta and Fonarow (2018) document declines in 30-day readmissions but stable 90-day readmissions and rising mortality—consistent with delayed readmission or premature discharge.  
- **Equity:** Joynt (2013) and Rogstad (2022) find that HRRP disproportionately penalized safety-net hospitals serving lower-income populations.  

This project builds on these strands by analyzing all three behavioral mechanisms within one unified framework and linking them directly to hospital penalty exposure.

---

## 3. Data and Variables

| Dataset | Content | Use |
|----------|----------|-----|
| **MedPAR (Inpatient)** | Admission, discharge, DRG codes, payments | Readmissions, DRG weights, LOS |
| **Outpatient File** | Observation claims (rev codes 0760–0769, G0378/9) | Observation utilization trends |
| **MBSF** | Demographics, comorbidities, death date | Mortality and control variables |
| **AHA Annual Survey** | Hospital ownership, bed size, teaching status | Structural covariates |
| **CMS HRRP Penalty File** | Annual hospital-level penalty rates | Exposure intensity |
| **DSH & Area Deprivation Index (ADI)** | Socioeconomic context | Equity heterogeneity |

---

## 4. Methods

### 4.1 Analytical Strategy

A **Difference-in-Differences (DiD)** and **Triple-Differences (DDD)** design will isolate the causal effects of HRRP across conditions and hospital types.

#### General model:
$$
Y_{hct} = \alpha + \beta(Post_t\times Target_c) + \mu_h + \lambda_t + \eta_c + \epsilon_{hct}
$$

where:
- \(Y_{hct}\): outcome (e.g., readmission rate, observation rate, mean DRG weight, mortality)  
- \(Post_t\): indicator for post-HRRP years (≥2013)  
- \(Target_c\): HRRP-targeted condition indicator  
- \(\mu_h\): hospital fixed effects  
- \(\lambda_t\): year fixed effects  
- \(\eta_c\): condition fixed effects  

**Triple-Differences (DDD):**
$$
Y_{hct} = \alpha + \theta(Post_t\times Target_c\times HighPenalty_h) + \text{FE} + \epsilon_{hct}
$$
This specification tests whether hospitals with greater HRRP exposure showed stronger behavioral responses.

---

### 4.2 Key Outcomes by Mechanism

| Mechanism | Variable | Expected Post-HRRP Change |
|------------|-----------|----------------------------|
| **Substitution** | Observation stays / total encounters | Ambiguous (empirical test) |
| **Upcoding** | Mean DRG weight, % with CC/MCC | ↑ (greater coding intensity) |
| **Timing** | Ratio of 31–90 day to 0–30 day readmissions; 30-day mortality | ↑ (delayed readmissions) |
| **Equity** | Differences by DSH quartile or ADI | ↑ behavioral intensity in low-SES hospitals |

---

### 4.3 Socioeconomic Adjustment

To test whether behavioral adaptations vary by socioeconomic status:

$$
Y_{hct} = \alpha + \beta_1(Post\times Target) + \beta_2(Post\times Target\times LowSES_h) + \text{FE} + \epsilon_{hct}
$$

If \(\beta_2>0\), HRRP effects were amplified among disadvantaged hospitals—indicating inequitable adaptation.

---

## 5. Empirical Plan and Timeline

| Phase | Timeframe | Objectives | Deliverables |
|-------|------------|-------------|--------------|
| **1. Data Assembly** | Weeks 1–3 | Merge MedPAR, Outpatient, HRRP, DSH, and AHA data; define hospital-year panel | Clean dataset & variable codebook |
| **2. Descriptive Analysis** | Weeks 4–6 | Plot readmissions, observation, DRG weights, and mortality trends | Diagnostic figures (2008–2020) |
| **3. Mechanism Testing (DiD)** | Weeks 7–10 | Estimate effects for each mechanism | Regression results & robustness |
| **4. SES Heterogeneity** | Weeks 11–12 | Add DSH/ADI interactions | Stratified results table |
| **5. Draft & Policy Implications** | Weeks 13–14 | Integrate findings, discuss reform implications | Research paper draft & 2-page memo |

---

## 6. Expected Contributions

1. **Empirical clarification:** Provides the first systematic test showing that observation-stay substitution may not explain HRRP’s effects.  
2. **Behavioral decomposition:** Quantifies three distinct mechanisms—substitution, upcoding, and timing—to explain aggregate trends.  
3. **Equity evaluation:** Examines whether hospitals serving vulnerable populations adapted differently under identical incentives.  
4. **Policy relevance:** Offers evidence-based guidance for refining HRRP’s design—suggesting closer monitoring of coding practices and mortality consequences.

---

## 7. Policy Implications

- **Broaden measurement:** Incorporate observation stays and post-30-day events into readmission metrics.  
- **Strengthen monitoring:** Track DRG-weight inflation to detect administrative gaming.  
- **Protect equity:** Adjust HRRP penalties for social risk and hospital resource constraints.  
- **Refocus incentives:** Tie rewards to total patient outcomes rather than narrow metrics.

---
```
