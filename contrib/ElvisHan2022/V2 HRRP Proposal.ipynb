# Hospitals’ Substitution under HRRP: Patient Financial Burden, Access, and Quality Effects

## 1. Introduction

Value-based payment reforms aim to realign provider incentives away from volume and toward value. The **Hospital Readmissions Reduction Program (HRRP)** penalizes hospitals with excess 30-day readmissions for targeted conditions, intending to improve care coordination and reduce preventable rehospitalizations.  

However, when penalties apply only to inpatient readmissions, hospitals can **reclassify returning patients as observation stays** billed under Medicare Part B. This behavior lowers measured readmissions but can **shift costs to patients**, **reduce access to post-acute care**, and **fragment care continuity**.  

This study evaluates the **costs and benefits** of HRRP by quantifying the extent of **observation substitution** and assessing its downstream effects on patient welfare, quality, and financial burden.

---

## 2. Literature Review

### 2.1 Overview

The literature on HRRP examines how financial incentives affect hospital behavior, quality, and efficiency. It focuses on four questions:

1. How did HRRP change **readmissions**, **observation stays**, and **ED visits**?  
2. Did these changes affect **mortality**, **continuity**, and **patient outcomes**?  
3. How did effects differ by **hospital type** and **social risk**?  
4. Did HRRP generate **net value** when accounting for patient costs?

---

### 2.2 Readmissions, Observation, and ED Utilization

Empirical studies show that HRRP **reduced inpatient readmissions** but caused **increases in observation stays and ED visits**, reflecting hospitals’ strategic response to avoid penalties.  

- *Sheehy (2021)*, *Wright (2023)*, and *Albritton (2017, 2018)* document large shifts toward observation care.  
- *Gupta (2018, 2021)* and *Catalyst (2018)* confirm system-wide substitution patterns.

**Takeaway:** HRRP achieved its numerical goal but induced **reclassification behavior**, which may not reflect true quality improvement.

---

### 2.3 Mortality and Clinical Quality

Evidence on mortality is **mixed**.  
- *Gupta (2017)*, *Fonarow (2017)*, and *Qiu (2018)* report higher mortality for heart failure post-HRRP.  
- *Murray (2021)* and *Samarghandi (2019)* attribute these findings to confounding by patient demographics and risk selection.  

Beyond mortality, studies note higher **ED revisits**, **longer observation durations**, and **shorter inpatient stays**, suggesting reduced continuity of care.

**Takeaway:** Mortality alone underestimates HRRP’s impact; **continuity metrics** like ED revisits and follow-up rates better capture post-policy quality.

---

### 2.4 Financial Burden and Access

Observation billing under **Part B** increases **out-of-pocket (OOP)** liabilities and prevents Medicare coverage for **Skilled Nursing Facility (SNF)** stays under the “three-day rule.”  
While *Gupta (2018; 2021)* and *Catalyst (2018)* discuss these mechanisms, few quantify them, leaving the **patient welfare impact** largely unmeasured.

---

### 2.5 Heterogeneity and Equity

- *Rogstad (2022)* and *Evans (2021)* highlight that social determinants and socioeconomic risk factors drive readmission rates.  
- *Kirsch (2024)* finds HRRP penalties fall disproportionately on **safety-net hospitals** and clinicians, potentially widening equity gaps.

---

### 2.6 Economic Value and Measurement

The program’s efficiency is debated:  
- *Yakusheva (2020)* and *Nuckols (2017)* find limited net savings after considering offsetting costs.  
- *Sheehy (2023)* critiques HRRP’s **measurement design**, arguing that it rewards coding behavior more than genuine value creation.  
- *Rothstein (2009)* similarly shows that incentive programs can shift costs without improving welfare.

**Takeaway:** Evaluating HRRP requires a **cost–benefit framework** balancing system savings against patient burdens.

---

### 2.7 Synthesis and Gaps

| Dimension | Consensus | Remaining Gap |
|------------|------------|----------------|
| Readmissions ↓, Observation ↑ | Strong | No index quantifying substitution intensity |
| Mortality & Quality | Mixed | Need multidimensional metrics |
| Patient Burden | Recognized | Sparse quantitative evidence |
| Equity | Important | Limited causal identification |
| Economic Value | Unclear | No welfare accounting model |

**Summary:**  
Existing studies confirm that hospitals substituted observation for readmissions but rarely quantify **patient-level financial and access harms**. This project fills that gap using a hospital-level **Observation Substitution Index (OSI)** linked to outcomes such as OOP, SNF coverage, and ED churn.

## 3. Conceptual Framework

Under HRRP (implemented FY2013), hospitals faced direct penalties for readmissions but none for observation care. This asymmetric incentive led to **strategic substitution**:

$$
\text{HRRP Penalties} \Rightarrow \text{↓ Readmissions, ↑ Observation Stays}
$$

This shift has several potential **downstream effects**:

- Higher **OOP spending** (Part B billing)  
- Reduced **SNF eligibility** (three-day rule)  
- Increased **ED/OBS churn** and delayed follow-up  
- Possible increases in **mortality** or missed admissions

---

## 4. Hypotheses

1. **Observation Substitution:**  
   Hospitals with higher OSI values experience greater post-HRRP increases in patient OOP spending.  

2. **Access:**  
   High-OSI hospitals show larger declines in covered SNF admissions.  

3. **Quality:**  
   High-OSI hospitals exhibit higher ED revisits and longer observation durations, reflecting degraded continuity.

---

## 5. Research Design

### 5.1 Observation Substitution Index (OSI)

Define OSI as the ratio of a hospital’s change in observation rate to its reduction in readmission rate:

$$
OSI_h = \frac{\Delta \text{Observation Rate}_h}{|\Delta \text{Readmission Rate}_h|}
$$

where $\Delta$ represents the change between pre- (2008–2012) and post-HRRP (2013–2019) periods.

---

### 5.2 Identification Strategy

#### Difference-in-Differences (DiD)

$$
Y_{iht} = \alpha + \beta (Post_t \times HighOSI_h) + \gamma X_{iht} + \mu_h + \lambda_t + \epsilon_{iht}
$$

- \(Y_{iht}\): patient outcome (OOP, SNF, ED revisit, etc.)  
- \(HighOSI_h\): indicator for hospitals in the top OSI quartile  
- \(Post_t\): post-HRRP period indicator  
- \(X_{iht}\): controls (age, sex, comorbidities, dual eligibility, hospital size/type)  
- \(\mu_h\): hospital fixed effects; \(\lambda_t\): year fixed effects

#### Triple Difference (DDD)

$$
Y_{iht} = \alpha + \theta (Post_t \times HighOSI_h \times Targeted_c) + \text{controls} + \mu_h + \lambda_t + \eta_c + \epsilon_{iht}
$$

where \(Targeted_c = 1\) for HRRP-targeted conditions (Heart Failure, AMI, Pneumonia, COPD, Hip/Knee).

---

## 6. Outcomes

### Financial
- Mean and 90th percentile **30-day OOP spending**  
- Probability of **catastrophic OOP** (> \$500 / \$1,000)

### Access and Continuity
- **SNF admission** within 7 days  
- **Covered SNF share**  
- **ED revisit** within 7/30 days  
- **Observation duration** >24h or >48h  
- **Follow-up** with PCP/specialist within 14 days

### System Efficiency
- **30-day readmission rate**  
- **Inpatient bed-days per 100 discharges**  
- **Program spending** (Part A + B)

### Optional Health Outcome
- **30-/90-day mortality**

## 8. Proposed Studies

### Study 1: Incentive-Driven Substitution and Patient Burden

**Design:**  
Difference-in-Differences comparing High vs. Low OSI hospitals, 2008–2019, with event-study diagnostics.  
Triple-Difference adds condition targeting.

**Primary Outcomes:**  
OOP spending, SNF coverage, ED revisits, observation duration, and readmissions.  

**Expected Findings:**  
High-OSI hospitals show larger post-HRRP increases in OOP and access frictions, particularly for targeted conditions.

---

### Study 2: Welfare Ledger — Quantifying Costs and Benefits

**Objective:**  
Translate regression estimates into a per-100-discharge cost–benefit ledger.

**Benefits (System Side):**  
- Fewer readmissions → **inpatient bed-days saved**  
- Possible **reductions in Medicare Part A spending**

**Costs (Patient Side):**  
- Increased **OOP**  
- **Lost SNF coverage**  
- More **ED/OBS churn**

**Deliverable:**  
Condition-specific **welfare tables** (Cardiovascular, Pulmonary, Orthopedic) showing net welfare impact.

---

## 9. Expected Contribution

This study reframes HRRP through a **value-based care lens**, linking provider behavior to **patient welfare**.  
It contributes by:

- Creating a **behavioral exposure (OSI)** that quantifies substitution intensity;  
- Combining **clinical**, **financial**, and **access** outcomes in one welfare framework;  
- Providing empirical evidence to inform **HRRP reform**, such as including observation stays in penalties or revising SNF eligibility.
