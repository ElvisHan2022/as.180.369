```python
# Hospitals’ Behavioral Adaptation under the Hospital Readmissions Reduction Program (HRRP)

**Author:** Elvis Han  
**Advisors:** Prof. Seth Richards-Shubik & Prof. Chris Carroll  
**Environment:** Jupyter / SciServer  
**Date:** 10/20/2025  

---

## 1. Motivation and Background

The **Hospital Readmissions Reduction Program (HRRP)**, implemented by the Centers for Medicare and Medicaid Services (CMS) in 2013, penalizes hospitals with excess 30-day readmissions for selected conditions.  
While the policy aimed to improve care coordination and reduce preventable rehospitalizations, emerging research suggests that hospitals may have **optimized behaviorally rather than clinically** — through administrative adjustments such as observation substitution, upcoding, and timing manipulation.

This proposal outlines a data-driven exploratory and causal analysis plan that investigates whether HRRP produced genuine efficiency gains or merely reshaped reporting and classification behavior. It links **incentive exposure** (penalties) to **behavioral adaptation** (hospital responses) and finally to **welfare outcomes** (patient costs, post-acute access, and quality).

---

## 2. Literature Insight: Gupta & Fonarow (2018)

### 2.1 Key Findings

Gupta & Fonarow (2018), *The Hospital Readmissions Reduction Program — Learning from Failure of a Healthcare Policy*, highlighted several critical points:

- **Limited success:** HRRP reduced readmissions modestly (≈9%), far less than the expected 25%.  
- **Gaming behaviors:** Hospitals adapted by:  
  1. **Upcoding:** Raising recorded severity to inflate expected readmissions.  
  2. **Observation substitution:** Reclassifying readmissions as outpatient “observation stays.”  
  3. **Timing manipulation:** Delaying rehospitalizations beyond 30 days.  
- **Equity problem:** Safety-net hospitals faced disproportionate penalties due to lack of socioeconomic risk adjustment.  
- **Mortality effects:** Rising heart-failure mortality post-HRRP suggested unintended harm.  
- **Policy lesson:** Financial incentives can misfire when performance metrics are narrow or misaligned.

### 2.2 Impact on Research Direction

Initially, this project aimed to test Gupta’s **observation substitution** hypothesis using outpatient data. However, preliminary exploratory analysis revealed that observation stays **did not rise** post-2013 — contradicting expectations.  

This “null result” reshaped the approach into a broader behavioral framework examining three potential mechanisms:

| Mechanism | Behavioral Channel | New Data/Index Introduced |
|------------|--------------------|---------------------------|
| **Observation substitution** | Reclassifying inpatient returns as outpatient observation | **Observation Substitution Index (OSI)** |
| **Upcoding** | Inflating coded severity or expected readmissions | **Coding Intensity Index** |
| **Timing manipulation** | Delaying readmissions beyond 30 days | **Timing Manipulation Index (Bunching)** |

This insight also motivated exploration of **equity heterogeneity** — testing whether resource-rich hospitals (teaching, urban, for-profit) had greater capacity to adapt strategically than safety-net or rural hospitals.

---

## 3. The Tension Between Administrative Aggregates and Service-Level Data

### 3.1 Conceptual Background

In evaluating HRRP, one of the most significant practical challenges lies in reconciling **aggregate administrative data** (system-level trends) with **micro-level provider data** (hospital behavior).  
The two data types represent fundamentally different lenses:

- **Medicare Summary (CPS A/B):** National spending and utilization aggregates under Part A (inpatient) and Part B (outpatient). It reflects **system-wide shifts** but masks local variation.  
- **Provider and Service-Level Data:** Detailed hospital-level discharge, charge, and DRG records capturing **behavioral responses** at the point of care.

These datasets reveal different truths:  
- The **CPS summary** tells *what happened overall* — a shift from inpatient to outpatient spending.  
- The **provider-level data** tell *how hospitals individually behaved* — who coded differently, delayed readmissions, or faced penalties.

---

### 3.2 Comparison and Trade-offs

| Dimension | Medicare Summary (CPS A/B) | Provider-Level (Inpatient/Outpatient) |
|------------|----------------------------|---------------------------------------|
| **Unit of Observation** | Year / National | Provider × DRG × Year |
| **Analytical Level** | Population / Aggregate | Micro / Behavioral |
| **Strength** | Captures macro inflection after HRRP | Reveals heterogeneity and incentive sensitivity |
| **Limitation** | Masks variation by hospital type | Lacks full spending and policy context |
| **Best Use** | Contextual validation of HRRP impact | Causal analysis of behavioral adaptation |

---

### 3.3 Empirical Tension Observed

In the data explored so far:

- **CPS Medicare Summary (A/B)** shows a sharp **pivot after 2013**, with Part A spending flattening and Part B spending rising — consistent with national-level substitution from inpatient to outpatient care.  
- **Provider-Level Data**, however, reveal **no corresponding drop** in discharges or case counts. Instead, hospitals report **rising per-case charges and increasing severity mix**, hinting at *administrative adaptation rather than clinical substitution*.

This contradiction raises key interpretive questions:  
- Did hospitals *actually shift* utilization patterns (true substitution)?  
- Or did they *relabel* and *recode* activity (strategic reporting)?

Understanding this mismatch is central to distinguishing **real efficiency gains** from **metric optimization**.

---

### 3.4 Integrative Strategy

Rather than choosing between datasets, the project integrates both:

1. **Macro Validation (CPS A/B):** Identify system-wide inflection points and policy timing effects.  
   - *Visualization:* Line chart of Part A vs. Part B payments (2008–2021).  
   - *Interpretation:* Confirms whether HRRP coincided with structural change.

2. **Micro Causal Analysis (Provider-Level):**  
   Quantify whether hospital penalties correspond with shifts in readmissions, observation stays, or coding intensity.  
   - *Visualization:* Scatterplots of ERR vs. Δcharges or OSI; event-study trends around 2013.  

3. **Crosswalk / Synthesis:**  
   Compare whether the magnitude of hospital-level substitution can account for national trends.  
   - If not, remaining variation may stem from upcoding or reporting shifts.

This multi-level approach transforms the tension between datasets into a **triangulation strategy** — using each to validate or challenge the other.

---

## 4. Empirical Framework and Identification Strategy

### 4.1 Conceptual Overview

HRRP provides a quasi-experimental setting suitable for **Difference-in-Differences (DiD)** analysis.  
Hospitals penalized under HRRP faced stronger incentives to reduce readmissions, while non-penalized hospitals serve as a control group.  
A **Triple-Differences (DDD)** extension further contrasts targeted and non-targeted conditions.

> **Research Question:**  
> Did HRRP’s penalties causally reduce readmissions, or did hospitals adapt by reallocating utilization across unpenalized categories (observation, coding, or timing)?

---

### 4.2 Baseline Difference-in-Differences Model

Let:

- \( Y_{ht} \): outcome for hospital \( h \) in year \( t \) (e.g., readmission rate, observation rate).  
- \( \text{Post}_t \): indicator for post-HRRP years (2013 onward).  
- \( \text{Penalized}_h \): 1 if hospital was penalized in first HRRP year.  
- \( X_{ht} \): control covariates (ownership, teaching status, DSH, region, size).  
- \( \mu_h, \lambda_t \): hospital and year fixed effects.  

Model:

$$
Y_{ht} = \alpha + \beta (\text{Post}_t \times \text{Penalized}_h) + \gamma X_{ht} + \mu_h + \lambda_t + \varepsilon_{ht}
$$

- \( \beta \): difference in outcomes among penalized hospitals post-HRRP.  
  - \( \beta < 0 \): real quality improvement (fewer readmissions).  
  - \( \beta > 0 \): evidence of behavioral adaptation (e.g., more observation coding).

---

### 4.3 Triple-Differences (DDD) Extension

For targeted vs. non-targeted conditions:

$$
Y_{hct} = \alpha + \theta (\text{Post}_t \times \text{Penalized}_h \times \text{Target}_c) + \mu_h + \lambda_t + \eta_c + \varepsilon_{hct}
$$

- \( \eta_c \): condition fixed effects.  
- \( \theta \): identifies HRRP’s condition-specific causal effect.

Interpretation:
- \( \theta < 0 \): true reduction in penalized-condition readmissions.  
- \( \theta > 0 \): substitution or recoding behavior.

---

### 4.4 Event-Study Specification

$$
Y_{ht} = \alpha + \sum_{k \neq -1} \beta_k D_{t=k} \times \text{Penalized}_h + \mu_h + \lambda_t + \varepsilon_{ht}
$$

- \( D_{t=k} \): years relative to HRRP implementation.  
- Pre-period (\( k<0 \)): tests for parallel trends.  
- Post-period (\( k>0 \)): tracks dynamic response after HRRP.

---

### 4.5 Outcome Variables

| Outcome \(Y\) | Mechanism | Expected Sign | Interpretation |
|----------------|------------|----------------|----------------|
| Readmission rate | Direct effect | \( \beta, \theta < 0 \) | Quality improvement |
| Observation rate | Substitution | \( \beta, \theta > 0 \) | Reporting shift |
| Severity mix | Upcoding | \( \beta, \theta > 0 \) | Coding inflation |
| Returns after 30 days | Timing manipulation | \( \beta, \theta > 0 \) | Delayed readmissions |
| SNF rehospitalization | Welfare | \( \beta, \theta > 0 \) | Patient-level cost or access trade-off |

---

## 5. Exploratory Data Analysis (EDA) Design

EDA examines each link in the chain:  
**Policy → Incentive → Behavior → Welfare**

---

### 5.1 Macro Context: Inpatient vs. Outpatient Shifts

**Goal:** Confirm macro substitution from inpatient (Part A) to outpatient (Part B).  
**Dataset:** CPS Medicare Summary (A/B).  
**Visualizations:**
- Line chart: Part A vs. Part B payments (2008–2021).  
- Stacked area: Spending shares by Part.  
- Annotated HRRP launch (2013).  
**Interpretation:** System-wide substitution pattern validates incentive reallocation.

---

### 5.2 Incentive Exposure: Penalty Variation

**Dataset:** HRRP Supplemental File.  
**Visualizations:**  
- Histogram: ERR distribution.  
- Scatter: ERR vs. Penalty factor.  
- Line: % penalized hospitals by year.  
- Heatmap: ERR by condition × year.  
**Goal:** Show heterogeneity in penalty pressure enabling DiD identification.

---

### 5.3 Behavioral Response: Substitution, Upcoding, Timing

#### Observation Substitution
- **Measure:** \( OSI = \frac{\Delta Observation}{|\Delta Readmission|} \)
- **Visuals:** Scatterplots (ΔObs vs. ΔReadm), OSI by penalty quartile, state-level choropleths.  

#### Upcoding
- **Measure:** Change in share of DRGs with CC/MCC.
- **Visuals:** Severity-mix charts, ΔSeverity vs. ΔLOS, boxplots by ownership.  

#### Timing Manipulation
- **Measure:** Ratio of returns (31–36 days)/(25–30 days).
- **Visuals:** Bunching histograms, event-study around 30-day cutoff.

---

### 5.4 Welfare Outcomes: SNF and Patient-Level Impact

**Dataset:** Nursing Home Claims-Based Quality.  
**Visualizations:**  
- Line chart: SNF rehospitalization & community discharge trends.  
- Scatter: OSI vs. SNF rehosp.  
- Boxplot: SNF quality by OSI quartile.  
**Goal:** Link behavioral adaptation to patient welfare.

---

## 6. Analytical Roadmap

| Stage | Objective | Output |
|--------|------------|--------|
| Stage 1 | Validate pre-trends | Event-study plots |
| Stage 2 | Estimate DiD (β) | Policy effect on hospital behavior |
| Stage 3 | Extend to DDD (θ) | Condition-specific responses |
| Stage 4 | Link to welfare | Regression results with SNF data |
| Stage 5 | Robustness | Sensitivity and heterogeneity checks |

---

## 7. Expected Contributions

- **Empirical:** Documents heterogeneous hospital adaptation under HRRP.  
- **Methodological:** Integrates macro and micro datasets into a unified DiD/DDD framework.  
- **Policy:** Evaluates whether HRRP improved quality or merely shifted costs administratively.  
- **Equity:** Examines whether resource disparities shaped hospitals’ ability to “game” penalties.

---

**Summary Thought:**  
This project converts the tension between datasets into a strength — using **macro evidence to frame context** and **micro data to uncover behavior**.  
Even if observation substitution proves weaker than expected, triangulating substitution, upcoding, and timing manipulation exposes the full behavioral landscape of HRRP.  
Ultimately, the study contributes to understanding how financial incentives ripple through hospital decision-making — revealing whether apparent efficiency gains reflect *real care improvement* or *statistical illusion*.
```
